***String Cleaning Function:***

````python
def clean_string(val):
    # Cleans categorical strings by:
    # Converting to string
    # Replacing known bad separators (@, _, -, &, /) with space
    # Lowercasing
    # Removing extra whitespace
    if pd.isna(val):
        return val

    val = str(val).lower() # Converts to string + lowercase
    for sep in ['@', '_', '-', '&', '/', ":", ";", ","]:
        val = val.replace(sep, ' ') # Replaces known bad separators (@, _, -, &, /) with space
    
    val = val.strip()
    while '  ' in val:
        val = val.replace('  ', ' ')  # collapse multiple spaces
    return val
````

````python
# PURPOSE:
# The 'orders' table contains a customer_id column, which is only a technical identifier per order.
# However, a single real-world customer may appear in multiple orders with different customer_ids.
# Therefore, we wish to replace 'customer_id' with the stable identifier 'customer_unique_id',
# which allows us to analyze unique customers. This will be done in 4 steps.

# We perform a merge with the 'customers' table to retrieve the matching 'customer_unique_id'
# for each order. Once added, we no longer need the 'customer_id', and we can drop it.

# STEP 1: Merge orders with customers to add customer_unique_id to each order
orders_df = orders_df.merge(
    customers_df[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)

# STEP 2: Drop the technical customer_id as we now use customer_unique_id instead
orders_df.drop(columns=['customer_id'], inplace=True)

# STEP 3: (Optional) Reorder columns to place customer_unique_id at the front
cols = ['order_id'] + ['customer_unique_id'] + [col for col in orders_df.columns if col != 'customer_unique_id' and col != 'order_id']
orders_df = orders_df[cols]

# STEP 4: Check the structure and preview the updated dataframe
orders_df.info()
orders_df.head()
````

````python
before = len(customers_df)
# Remove duplicate customer records based on customer's TRUE identifier = 'customer_unique_id'
# So that only the first occurrence is kept; the rest are dropped:
customers_df = customers_df.drop_duplicates(subset='customer_unique_id', keep='first')
customers_df.reset_index(drop=True, inplace=True)

# Drop rows where the customer_unique_id is missing (NaN) - this makes joining & drawing conclusions on customers impossible!
# customers_df = customers_df.dropna(subset=['customer_id'])
after = len(customers_df)
print(f"Removed {before - after} duplicate customers.") # Log number of duplicates removed
customers_df.info() # checking for basic info about the customers table - number of columns, rows, column names & types, etc..
````

````python
# Looking for payments with negative payment_value (illogical, shouldn't happen):
negative_payment_values = payments_df["payment_value"] < 0

print("Payments with negative payment_value:")
print(negative_payment_values.sum())  # Printing the number of such records
print("\n")
print(negative_payment_values.mean())  # Printing the proportion of such records

# A payment value should always be non-negative - negative payments indicate invalid transactions.
# These records are likely corrupt data and thus removed.
# Furthermore, these errors don't appear a lot and so removing them won't change much.
# Also, replacing the errors with a constant value might inject bias.
payments_df = payments_df[payments_df["payment_value"] >= 0]
````

````python
# Looking for payments with negative installment values:
negative_installments = payments_df["payment_installments"] < 0

print("Payments with non-positive number of installments (0 or less):")
print(negative_installments.sum())  # Printing the number of such records
print("\n")
print(negative_installments.mean())  # Printing the proportion of such records
      
# Installments represent the number of monthly payments. We remove such rows for a few reasons:
# 1. Values <= 0 are illogical and likely data errors.
# 2. These errors don't appear a lot and so removing them won't change much.
# 3. Replacing the errors with a constant value might inject bias.
payments_df = payments_df[payments_df["payment_installments"] >= 0]
````

````python
# Looking for payments with non-positive installment values:
negative_sequentials = payments_df["payment_sequential"] < 0

print("Payments with non-positive number of sequentials (0 or less):")
print(negative_sequentials.sum())  # Printing the number of such records
print("\n")
print(negative_sequentials.mean())  # Printing the proportion of such records
      
# Payment numbers represent the number of the payment out of the defined number of payments. We remove such rows for a few reasons:
# 1. Values < 0 are illogical and likely data errors.
# 2. These errors don't appear a lot and so removing them won't change much.
# 3. Replacing the errors with a constant value might inject bias.
payments_df = payments_df[payments_df["payment_installments"] >= 0]
````

````python
# Looking for payments with payment_sequential > number of payment_installments (invalid):
invalid_sequential = payments_df["payment_sequential"] > payments_df["payment_installments"]

print("Payments where sequential number > total installments:")
print(invalid_sequential.sum())  # Printing the number of such records
print("\n")
print(invalid_sequential.mean())  # Printing the proportion of such records

# A payment_sequential should never be larger than the number of payment_installments - this would indicate invalid payments.
# These records are likely corrupt data and thus removed.
# Furthermore, these errors don't appear a lot and so removing them won't change much.
# Also, replacing the errors with a constant value might inject bias.
payments_df = payments_df[payments_df["payment_sequential"] <= payments_df["payment_installments"]]
````

#### Converting all dates to datetime format & defining the Dim-Date dimension table:

````python
# Create a date dimension table (dim_date) from all unique date fields across all datasets
# This table helps simplify time-based queries and supports standard time hierarchies (day, week, month, etc.)
# We shall create it in steps:

# Step 1: Convert all date columns across relevant datasets
# (We use 'errors="coerce"' to safely convert problematic values to NaT)
date_columns_orders = [
    'order_purchase_timestamp', 'order_approved_at',
    'order_delivered_carrier_date', 'order_delivered_customer_date',
    'order_estimated_delivery_date'
    ]

for col in date_columns_orders:
    orders_df[col] = pd.to_datetime(orders_df[col], errors='coerce')
    
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'], errors='coerce')
sellers_df['registration_date'] = pd.to_datetime(sellers_df['registration_date'], errors='coerce')
customers_df['registration_date'] = pd.to_datetime(customers_df['registration_date'], errors='coerce')

if 'transaction_date' in payments_df.columns:
    payments_df['transaction_date'] = pd.to_datetime(payments_df['transaction_date'], errors='coerce')

    
# Step 2: Collect all unique datetime values from the various columns:
all_dates = pd.Series(pd.concat([
    orders_df[col] for col in date_columns_orders
] + [
    order_items_df['shipping_limit_date'],
    sellers_df['registration_date'],
    customers_df['registration_date'],
    payments_df['transaction_date'] if 'transaction_date' in payments_df.columns else pd.Series(dtype='datetime64[ns]')
], axis=0).dropna().unique()).drop_duplicates()

# Step 3: Create the actual date dimension (dim_date):
dim_date = pd.DataFrame({'date': pd.to_datetime(all_dates)})
dim_date['day'] = dim_date['date'].dt.day
dim_date['weekday'] = dim_date['date'].dt.weekday
dim_date['month'] = dim_date['date'].dt.month
dim_date['year'] = dim_date['date'].dt.year
dim_date['quarter'] = dim_date['date'].dt.quarter
dim_date['is_weekend'] = dim_date['weekday'].isin([5, 6])  # Saturday or Sunday

# The result can be joined by date for consistent time-based analytics
dim_date.reset_index(drop=True, inplace=True)

# Step 5: Remove duplicates & NAs from dim_date:
# Additional duplicate removal into dim_date:
dim_date = dim_date.drop_duplicates(subset=['date'])

# Additional NA removal in dim_date:
dim_date = dim_date[dim_date['date'].notna()]

dim_date.info() # checking for basic info about the dates table - number of columns, rows, column names & types, etc..
dim_date.head(50)
````

#### Handling Date Inconsistencies - primarily on orders table but also customers, sellers, products & payments table:

The goal here is to assess logical inconsistencies in date-related fields in the aformentioned tables, and decide how best to handle them (remove, flag, impute, etc.) depending on the nature and severity of the issue.
For each case below, we'll compute the number and proportion of such cases, and explain the suggested handling strategy.

*Prerequisite:* ensure all relevant columns are in datetime format

````python
# 1. Orders with status = 'shipped' but missing 'order_delivered_carrier_date'
shipped_no_shipping = orders_df[(orders_df['order_status'] == 'shipped') & (orders_df['order_delivered_carrier_date'].isna())]
print("Orders with status='shipped' but no shipping date:")
print(shipped_no_shipping.shape[0]) # Printing the number of such records
print("\n")
print(shipped_no_shipping.shape[0] / len(orders_df)) # Printing the proportion of such records
# Handling:  Let them stay as NaT but can be flagged when analyzing timelines. No imputation is done to preserve integrity.
orders_df.loc[shipped_no_shipping.index, 'order_delivered_carrier_date'] = pd.NaT
````

````python
# 2. Shipped orders missing shipping *or* purchase/approval dates
shipped_inconsistent = orders_df[(orders_df['order_status'] == 'shipped') & (
    orders_df['order_delivered_carrier_date'].isna() |
    orders_df['order_purchase_timestamp'].isna() |
    orders_df['order_approved_at'].isna()
)]
print("Shipped orders with missing shipping/purchase/approval dates:")
print(shipped_inconsistent.shape[0]) # Printing the number of such records
print("\n")
print(shipped_inconsistent.shape[0] / len(orders_df)) # Printing the proportion of such records

# Handling: Selectively apply NaT on the dates only where needed
shipped_inconsistent = orders_df['order_status'] == 'shipped'

# Selectively apply NaT only where needed:
orders_df.loc[shipped_inconsistent & orders_df['order_delivered_carrier_date'].isna(), 'order_delivered_carrier_date'] = pd.NaT
orders_df.loc[shipped_inconsistent & orders_df['order_purchase_timestamp'].isna(), 'order_purchase_timestamp'] = pd.NaT
orders_df.loc[shipped_inconsistent & orders_df['order_approved_at'].isna(), 'order_approved_at'] = pd.NaT
````

````python
# 3. Delivered orders missing actual delivery date or other key timestamps:
no_actual_delivery = orders_df[(orders_df['order_status'] == 'delivered') & (
    orders_df['order_delivered_customer_date'].isna() |
    orders_df['order_purchase_timestamp'].isna() |
    orders_df['order_approved_at'].isna()
)]
print("Delivered orders missing delivery/purchase/approval date:")
print(no_actual_delivery.shape[0]) # Printing the number of such records
print("\n")
print(no_actual_delivery.shape[0] / len(orders_df)) # Printing the proportion of such records

# Handling: Selectively apply NaT on the dates only where needed
delivered_inconsistent = orders_df['order_status'] == 'delivered'

# Selectively apply NaT only where missing
orders_df.loc[delivered_inconsistent & orders_df['order_delivered_customer_date'].isna(), 'order_delivered_customer_date'] = pd.NaT
orders_df.loc[delivered_inconsistent & orders_df['order_purchase_timestamp'].isna(), 'order_purchase_timestamp'] = pd.NaT
orders_df.loc[delivered_inconsistent & orders_df['order_approved_at'].isna(), 'order_approved_at'] = pd.NaT
````

````python
# 4. Orders with purchase/approval but wrong status (e.g., still processing):
status_mismatch = orders_df[(orders_df['order_purchase_timestamp'].notna()) &
                             (orders_df['order_approved_at'].notna()) &
                             (~orders_df['order_status'].isin(['approved', 'shipped', 'delivered']))]
print("Orders approved but with unexpected status:")
print(status_mismatch.shape[0]) # Printing the number of such records
print("\n")
print(status_mismatch.shape[0] / len(orders_df)) # Printing the proportion of such records

# Handling: These could be updated to 'approved'
orders_df.loc[status_mismatch.index, 'order_status'] = 'approved'
````

````python
# 5. Orders without status 'shipped' but a known shipping date:
status_inconsistency = orders_df[(orders_df['order_status'] != 'shipped') &
                                  (orders_df['order_status'] != 'delivered') &
                                  (orders_df['order_delivered_carrier_date'].notna())]
print("Orders not marked as 'shipped' but with actual shipping date present:")
print(status_inconsistency.shape[0]) # Printing the number of such records
print("\n")
print(status_inconsistency.shape[0] / len(orders_df)) # Printing the proportion of such records

# Handling: Likely mislabeled status, therefore updated status to 'shipped' if shipping date is reliable.
orders_df.loc[status_inconsistency.index, 'order_status'] = 'shipped'
````

````python
# 5b. Orders with status 'shipped' but a known delivery date:
shipped_with_delivery = orders_df[(orders_df['order_status'] == 'shipped') &
                                  (orders_df['order_delivered_customer_date'].notna())]
print("Orders marked as 'shipped' but with actual delivery date present:")
print(shipped_with_delivery.shape[0]) # Printing the number of such records
print("\n")
print(shipped_with_delivery.shape[0] / len(orders_df)) # Printing the proportion of such records


# Handling: Likely mislabeled status, therefore updated status to 'delivered' if delivery date is reliable.
orders_df.loc[shipped_with_delivery.index, 'order_status'] = 'delivered'
````

````python
# 6. Broken purchase flow:
broken_flow = orders_df[(orders_df['order_purchase_timestamp'] > orders_df['order_approved_at']) |
                        (orders_df['order_approved_at'] > orders_df['order_delivered_carrier_date']) |
                        (orders_df['order_delivered_carrier_date'] > orders_df['order_delivered_customer_date'])]
print("Orders with invalid chronological order of events:")
print(broken_flow.shape[0]) # Printing the number of such records
print("\n")
print(broken_flow.shape[0] / len(orders_df)) # Printing the proportion of such records

broken_flows = pd.DataFrame()
broken_flows['purchase > approval'] = orders_df['order_purchase_timestamp'] > orders_df['order_approved_at']
broken_flows['approval > carrier'] = orders_df['order_approved_at'] > orders_df['order_delivered_carrier_date']
broken_flows['carrier > delivered'] = orders_df['order_delivered_carrier_date'] > orders_df['order_delivered_customer_date']


# Handling: selectively set only the invalid timestamps to NaT to prevent problematic computations based on erroneous durations
for col, field in zip(['purchase > approval', 'approval > carrier', 'carrier > delivered'],
                      ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']):
    orders_df.loc[broken_flows[col], field] = pd.NaT
````

````python
# 7. Sellers mismatch - Sellers with sales before their registration
seller_mismatch = order_items_df.merge(orders_df[['order_id', 'order_purchase_timestamp']], on='order_id')
seller_mismatch = seller_mismatch.merge(sellers_df[['seller_id', 'registration_date']], on='seller_id')
seller_issue = seller_mismatch[seller_mismatch['order_purchase_timestamp'] < seller_mismatch['registration_date']]
print("Sellers selling before registration:")
print(seller_issue.shape[0]) # Printing the number of such records
print("\n")
print(seller_issue.shape[0] / len(sellers_df)) # Printing the proportion of such records

# Handling: Selectively remove registration dates of "bad sellers", but what are bad sellers?
# Compute proportion of bad records per seller:
bad_seller_ratio = seller_issue.groupby('seller_id').size() / seller_mismatch.groupby('seller_id').size()
sellers_to_flag = bad_seller_ratio[bad_seller_ratio > 0.5].index  # e.g., over 50% suspicious

print("Sellers with >50% of sales before registration:", len(sellers_to_flag))
sellers_df.loc[sellers_df['seller_id'].isin(sellers_to_flag), 'registration_date'] = pd.NaT
````

````python
# 8. Customers mismatch - Customers with orders before registration
customer_mismatch = orders_df.merge(customers_df[['customer_unique_id', 'registration_date']], on='customer_unique_id')
customer_issue = customer_mismatch[customer_mismatch['order_purchase_timestamp'] < customer_mismatch['registration_date']]
print("Customers ordering before registration:")
print(customer_issue.shape[0]) # Printing the number of such records
print("\n")
print(customer_issue.shape[0] / len(customers_df)) # Printing the proportion of such records

# Handling: Similar to above - # Handling: Selectively remove registration dates of "bad customers", but what are bad customers?
# Compute proportion of bad records per customer:
bad_customer_ratio = customer_issue.groupby('customer_unique_id').size() / customer_mismatch.groupby('customer_unique_id').size()
customers_to_flag = bad_customer_ratio[bad_customer_ratio > 0.5].index

print("Customers with >50% of orders before registration:", len(customers_to_flag))
customers_df.loc[customers_df['customer_unique_id'].isin(customers_to_flag), 'registration_date'] = pd.NaT
````

````python
# 9. Transactions before purchase/approval date
payment_issue = payments_df.merge(orders_df[['order_id', 'order_purchase_timestamp', 'order_approved_at']], on='order_id', how='left')
payment_issue = payment_issue[(payment_issue['transaction_date'] < payment_issue['order_purchase_timestamp']) |
                               (payment_issue['transaction_date'] < payment_issue['order_approved_at'])]
payment_issue_idx = payment_issue.index
print("Payments with transaction before purchase/approval:")
print(payment_issue.shape[0]) # Printing the number of such records
print("\n")
print(payment_issue.shape[0] / len(payments_df)) # Printing the proportion of such records


# Handling: Removing transaction date entirely due to this major inconsistency appearing throughout most payments:
payments_df.drop(columns = ['transaction_date'], inplace = True)

payments_df.info() # checking for basic info about the payments table - number of columns, rows, column names & types, etc..
````

## Creating the Star Schema - Fact Table - Order-Items & Dimensions:

The chosen Fact Table - **Order-Items**.
This table captures the most granular level of transactional data - each item sold in an order.
Each row corresponds to a unique combination of order_id and order_item_id.

We chose to represent each transaction in this granularity for a few key reasons:
- It includes foreign keys to dimensions: order_id, product_id, seller_id.
- Measures like price and freight_value are additive and represent the metrics we would aggregate.
- Chosen as fact because it's the lowest level of event granularity and enables detailed sales analysis.

Furthermore, we decided to add all relevant important metrics which are also on the order-level such as average-review-score, all relevant order dates, number of order payments, total price & freight price etc. We did so because most of these columns are both numeric and highly relevant, and so despite the issue of many, many duplications, we decided to store those columns in the facts table anyways..

In order to facilitate an easier analysis on products, payments, sellers & customers, we designated each of those as a dimension table. Payments specifically was designated as a dimension table to avoid duplicating payment methods in fact_order_items, which is an especially serious problem w.r.t to payments.

We also constructed a date dimension table as well, as a calendar dimension to support date-based slicing of metrics, for the following reasons:

- Enables consistent date-based analysis across fact and dimension tables.
- Can be joined on various date fields like purchase, delivery, and registration.
- Although connecting multiple date fields to a single dimension may seem to violate the star schema, it's common to join dim_date multiple times using aliases (one per date type) to maintain usability.

````python
# CHOSEN FACT TABLE - fact_order_items, which contains transaction-level granularity - seller, customer, order, product..
fact_order_items = order_items_df.copy()

# Merge in required dimensions: orders, products, sellers
fact_order_items = fact_order_items.merge(
    orders_df[['order_id', 'customer_unique_id', 'order_status', 'order_purchase_timestamp',
               'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date',
               'order_estimated_delivery_date', 'total_price', 'total_freight', 'average_review_score', 'num_payment_methods']],
    on='order_id', how='left'
)


# DIMENSION: CUSTOMERS
dim_customers = customers_df.copy()

# DIMENSION: SELLERS:
dim_sellers = sellers_df.copy()

# DIMENSION: PRODUCTS:
dim_products = products_df.copy()

# DIMENSION: PAYMENTS:
# Represents payment information per order. 
# This is treated as a dimension to avoid duplicating payment methods in fact_order_items.
dim_payments = payments_df.copy()
````

````python
def fix_datetime_with_default(df, col, default_date='1900-01-01'):
    df[col] = pd.to_datetime(df[col], errors='coerce')
    df[col] = df[col].fillna(pd.to_datetime(default_date))


fix_datetime_with_default(dim_customers, 'registration_date')
fix_datetime_with_default(dim_sellers, 'registration_date')
fix_datetime_with_default(dim_products, 'launch_date')
fix_datetime_with_default(dim_date, 'date')

for col in [
    'shipping_limit_date',
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]:
    fix_datetime_with_default(fact_order_items, col)
````

````python
def clean_bad_values(df):
    # Replaces problematic values (e.g., NaN, NaT, inf, '', 'nan', 'null', etc.) with Python None,
    # ensuring compatibility with MySQL insertion.
    # Works across all data types including strings, floats, ints, timestamps, and numpy objects.

    bad_strings = {'', ' ', 'nan', 'none', 'null', 'NaN', 'Nan', 'NULL', 'NAN', 'None'}

    def clean_cell(val):
        # Handle standard nulls and missing values
        if pd.isnull(val):
            return None
        # Handle numpy-specific objects (like np.float64) with problematic values
        if isinstance(val, np.generic):
            if np.isnan(val) or np.isinf(val):
                return None
        # Handle standard floats or ints with special cases
        if isinstance(val, (float, int)) and (val == float('inf') or val == float('-inf') or np.isnan(val)):
            return None
        # Handle strings that are bad
        if isinstance(val, str) and val.strip().lower() in bad_strings:
            return None
        return val

    for col in df.columns:
        df[col] = df[col].apply(clean_cell)

    return df


# Apply to all relevant DataFrames before inserting to MySQL:
dim_customers = clean_bad_values(dim_customers)
dim_products = clean_bad_values(dim_products)
dim_sellers = clean_bad_values(dim_sellers)
dim_date = clean_bad_values(dim_date)
````

````python
def force_clean_dataframe(df):
    # Define bad string values
    bad_strings = {'', ' ', 'nan', 'none', 'null', 'NaN', 'Nan', 'NULL', 'NAN', 'None'}

    def clean_value(val):
        try:
            # NaNs or NaTs
            if pd.isnull(val):
                return None
            # Numpy types
            if isinstance(val, np.generic):
                if np.isnan(val) or np.isinf(val):
                    return None
            # Python native floats
            if isinstance(val, float) and (np.isnan(val) or np.isinf(val)):
                return None
            # Strings with bad values
            if isinstance(val, str) and val.strip().lower() in bad_strings:
                return None
        except:
            return None  # just in case something fails silently
        return val

    # Fully rebuild the DataFrame with safe values
    cleaned_data = [
        [clean_value(val) for val in row]
        for row in df.values
    ]
    return pd.DataFrame(cleaned_data, columns=df.columns)

# Apply to your table
fact_order_items = force_clean_dataframe(fact_order_items)
````
